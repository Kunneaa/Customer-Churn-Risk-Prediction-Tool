# Problem Framing

## Business Context
Customer churn impacts recurring revenue and long-term growth in subscription businesses.

## Goal
- Predict churn probability for each customer
- Segment risk into Low / Medium / High
- Support retention prioritization

## Data Split Strategy (Given by Dataset)
- `train.csv` is the training dataset with target `Churn`.
- `test.csv` is the hold-out dataset without target.
- Workflow is: Problem Framing -> EDA & Feature Engineering -> Model Benchmarking -> Evaluation/Submission.
- Model benchmark includes: Logistic Regression, Decision Tree, Random Forest, Naive Bayes, SVM Linear, SVM Poly, KNN, AdaBoost, XGBoost, CatBoost (if installed).
- Primary metric is ROC-AUC for imbalanced classification; average precision is reported as secondary support metric.
- EDA saves selected raw features to `models/selected_features_gsa.json` so training notebook can reuse them directly.
